```python
#Задание1 - загрузка данных
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv', na_values = '?')
df.head()
print(df.columns.tolist())
table = df[['2','1', '38.50', '66', '28', '3', '5', '2.2']]
table.head()
```

```python
table.columns = ['surgery', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']
```

```python
#Задание 2 - работа с данными, определение выбросов
table.describe(include = 'all')
```

```python

table['Age'].value_counts()
"""В описании данных возраст лошади может быть: 1 - взрослая лошадь и 2 -молодая (моложе 6 месяцев),
    в графе возраст 24 лошади имеют возраст 9 - выброс"""
table_age = table.loc[table ['Age'] == 9]
table_age


"""Рассмотрев другие показатели лошадей с категорией возраста 9, можно сделать вывод, 
что у всех этих лошадей пульсб температура и дыхание не соответсвует норме и половина из этих лошадей умерли, 
значит скорее всего эти лошади тоже взрослые и 9 в возрасте необходимо заменить на 1"""
```

```python
table_age['outcome'].value_counts()
```

```python
table['Age'] =table['Age'].replace(9, 1)
table
```

```python
# Задание 3 - работа с пропусками
table.isna().sum()
```

```python
""" Уберем столбцы, где больше одного пропуска"""
table2=table.dropna(thresh=7)
```

```python
table2.isna().sum()
```

```python
table2.describe()
```

```python

"""Пропуски температуры можно заполнить медианой, 
в зависимости от операций и исхода"""
print(table2.groupby(['surgery',  'outcome'])['rectal temperature'].median())
table2['rectal temperature'].fillna(table2.groupby(['surgery', 'outcome'])
                                    ['rectal temperature'].transform('median'), inplace = True)
```

```python
table2.loc[table2 ['pulse'].isna()].head()
```

```python

"""Пропуски пульса заполняем медианой в зависимости от операции, боли и исхода"""
print(table2.groupby(['surgery',  'outcome', 'pain'])['pulse'].median())
table2['pulse'].fillna(table2.groupby(['surgery', 'outcome', 'pain'])
                                                 ['pulse'].transform('median'), inplace = True)

```

```python
"""пропуски частоты дыхания заменяем медианой, в зависимости от боли, исхода и операций"""
print(table2.groupby(['outcome',  'surgery', 'pain'])['respiratory rate'].median())
table2['respiratory rate'].fillna(table2.groupby(['outcome', 'surgery', 'pain'])
                                                 ['respiratory rate'].transform('median'), inplace = True)

```

```python
table2.isna().sum()
```

```python
"""Пропуски категориальных показателей заполняем надписью - no data"""
table2['temperature of extremities'].fillna('no data', inplace = True)
table2['pain'].fillna('no data', inplace = True)
table2['outcome'].fillna('no data', inplace = True)
```
