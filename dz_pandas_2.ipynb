```python
import pandas as pd
```

```python
#Задание 1

data_ratings = pd.read_csv('ratings.csv')
data_movie = pd.read_csv('movies.csv')

def class_rat(rating):
    
    if rating <= 2.0:
        return 'Low rating'
    elif rating <= 4.0:
        return 'Middle rating'
    elif rating == 4.5 or rating == 5.0:
        return 'High rating'
    
data_ratings['class'] = data_ratings['rating'].apply(class_rat)
data_ratings.join(data_movie['title'], on = 'movieId')        
```

```python
#задание 2
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
keysreq = pd.read_csv('keywords.csv')
keysreq
```

```python
def region(row):
    row = str(row).split()
    for key, value in geo_data.items():
        for word in row:
            if word in value:
                return (key)
    return "undefined"
        
keysreq['region'] = keysreq['keyword'].apply(region)
keysreq
keysreq[(keysreq['region'] == 'Северо-Запад')]
```

```python
# Задание 3
year = []
year = [str(i) for i in range (1950, 2011)]
print(year)
```

```python
movies = pd.read_csv('movies.csv')
rating =pd.read_csv('ratings.csv')
def production_year(row):
    for el in year:
        if el in  row:
            return el
   
    return '1990'

movies['year'] = movies['title'].apply(production_year)
movies
mv = movies.join(rating['rating'], on='movieId')


pd.DataFrame(mv.groupby(['year']).mean()['rating']).sort_values('rating', ascending = False).head()
```
