```python
#Уровень 0:
#Задание 1
#Посчитайте значение первой и второй производной функции: f(x) =x**5  + 4sin(2x) + cos(3x+3)
import sympy as sym
from sympy import *
x = sym.Symbol('x')




#В точке x = 1.
```

```python
f1 =diff((x**5  + 4*sin(2*x) + cos(3*x+3)), x)
```

```python
f1
```

```python
f1.subs(x,1)
```

```python
f2 = diff(f1)
f2.subs(x,1)
```

```python
f =  sym.sin(2*x + 1)**5
f
```

```python
f1 = diff(f, x)
f1
```

```python
f2 = diff(f1, x)
f2
```

```python

from sympy.plotting import plot

plot(f1, f2, ('x', -5, 5, 100))


     
```
