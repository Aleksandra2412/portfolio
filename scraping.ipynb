```python
import pandas as pd
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
def get_habr_articles(queries):
    habr  = pd.DataFrame() 
    url = 'https://habr.com/ru/all/' 
    for query in queries:
        params = {'q': query}
        req = requests.get(url=url, params=params) 
        soup = BeautifulSoup(req.text)
        
        posts = soup.find_all('div', class_='tm-article-snippet')
        for post in posts:
            date = post.find('span', 'tm-article-datetime-published').text
            title = post.find('h2', 'tm-article-snippet__title_h2').text
            link = 'https://habr.com' + post.find('a', 'tm-article-snippet__title-link').get('href')
            
            row = {'date': date, 'title': title, 'link': link}
            habr = pd.concat([habr, pd.DataFrame([row])])

    return habr.reset_index(drop=True)
    
get_habr_articles(['python', 'анализ'])
```
